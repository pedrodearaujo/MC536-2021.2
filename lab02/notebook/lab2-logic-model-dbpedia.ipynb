## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import requests
import json 

url = "http://dbpedia.org/data/Tent_of_Miracles_(novel).jsod"
response = requests.get(url)
response_json = response.json()
print(json.dumps(response_json, indent="\t"))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
author_url = response_json["d"]["results"][0]["http://dbpedia.org/ontology/author"]["__deferred"]["uri"]

print("The author of Tent of Miracles is: " + author_url)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
url = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept = "aspirin"
match = "exact"
limit = "10"

url = url .replace("{concept}", concept)\
          .replace("{match}", match)\
          .replace("{limit}", limit)

data = requests.get(url)
json_mi = data.json()

uri_mesh = json_mi[0]["resource"]
bar = str.rindex(uri_mesh, "/")

url = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = uri_mesh[bar+1:]

url = url.replace("{code}", code)

data = requests.get(url)
json_mid = data.json()

print(json.dumps(json_mid, indent="\t"))
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
url = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"

format = "json"
query = "Eiffel Tower"

url = url .replace("{format}", format)\
          .replace("{query}", query)

data = requests.get(url)
tower_ref = data.json()

print(json.dumps(tower_ref, indent="\t"))
```

```python
print(tower_ref[0]["osm_id"])
```

```python
url = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "way"
id = str(tower_ref[0]["osm_id"])
format = "json"

url = url .replace("{element_type}", element_type)\
          .replace("{id}", id)\
          .replace("{format}", format)

data = requests.get(url)
tower_way = data.json()

print(json.dumps(tower_way, indent=3))
```

```python
print(tower_way["elements"][0]["tags"]["wikipedia"])
```

```python
url = "http://dbpedia.org/data/{concept}.{format}"

concept = tower_way["elements"][0]["tags"]["wikipedia"] .replace(" ", "_")\
                                                        .replace("fr:", "")
format = "jsod"

url = url .replace("{concept}", concept)\
          .replace("{format}", format)

data = requests.get(url)
tower_dbpedia = data.json()

print(json.dumps(tower_dbpedia, indent="\t"))
```
